```python
import pandas as pd
from pprint import pprint
```

```python
data = pd.read_csv('datos/data_gapminder_experiment.csv', sep=';', decimal=",")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score 
```

```python
X = data.drop(['country', 'Date'], axis=1)
contries = data['country']
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
clusters = 5
```

```python
from sklearn.cluster import AgglomerativeClustering

hierarchical = AgglomerativeClustering(n_clusters=clusters)
labels = hierarchical.fit_predict(X)


plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='rainbow', alpha=0.7)
plt.colorbar(scatter, label='Cluster Label')
plt.title("Clustering Results on PCA-Reduced Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()

sil_coeff = silhouette_score(X, labels)
print("Silhouette Coefficient:", round(sil_coeff, 3))
```

```python
coutry_by_label = {}

for country, label in zip(contries, labels):
    if label not in coutry_by_label:
        coutry_by_label[label] = [country]
    else:
        coutry_by_label[label].append(country)
pprint(coutry_by_label)
```

```python
from sklearn.cluster import KMeans


kmeans = KMeans(n_clusters=clusters, random_state=0)
labels = kmeans.fit_predict(X)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='rainbow', alpha=0.7)
plt.colorbar(scatter, label='Cluster Label')
plt.title("Clustering Results on PCA-Reduced Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
sil_coeff = silhouette_score(X, labels)
print("Silhouette Coefficient:", round(sil_coeff, 3))
```

```python
coutry_by_label = {}

for country, label in zip(contries, labels):
    if label not in coutry_by_label:
        coutry_by_label[label] = [country]
    else:
        coutry_by_label[label].append(country)
pprint(coutry_by_label)
```

```python
from sklearn.cluster import DBSCAN


dbscan = DBSCAN(eps=0.00000001, min_samples=5)
labels = dbscan.fit_predict(X)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='rainbow', alpha=0.7)
plt.colorbar(scatter, label='Cluster Label')
plt.title("Clustering Results on PCA-Reduced Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```
